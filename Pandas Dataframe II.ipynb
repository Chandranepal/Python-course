### Rank values with .rank() method

```python
df=pd.read_csv("nba.csv").dropna(how="all")
df.Salary=df.Salary.fillna(0).astype("int")
```

```python
df
```

```python
df["Salary Rank"]=df["Salary"].rank(ascending=False).astype("int")
```

Looking into the dataframe, we came to know that two columns are in date and time format and Senior Management is in boolean category and Gender is also category.

```python
df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
```

```python
df["Senior Management"]=df["Senior Management"].astype("bool")
df["Gender"]=df["Gender"].astype("category")
```

```python
df
```

### Filtering a dataframe based on one condition or more than one

```python
req=df["Gender"]=="Male"
df[req]
```

```python
#this can be also one as:
df[df["Gender"]=="Male"]
```

```python
req1=df["Gender"]=="Male" #1
req2=df["Team"]=="Marketing" #1
df[req1 & req2]              #1 & 1 = 1
```

```python
df[(df["Gender"]=="Male") & (df["Team"]=="Marketing")]
```

```python
req1=df["Gender"]=="Male" 
req2=df["Team"]=="Marketing" 
req3=df["Start Date"]>"2011-06-01"
df[req1 & req2 & req3]              
```

```python
req1=df["Gender"]=="Male" 
req2=df["Team"]=="Marketing" 
req3=df["Start Date"]>"2011-06-01"
df[(req1 & req2) | req3]    
```

```python
req1=df["Senior Management"]==True
req2=df["Start Date"]<"1990-01-01"
df[req1 & req2]
```
